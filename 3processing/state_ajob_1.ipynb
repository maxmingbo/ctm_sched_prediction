### 统计完成在今天0-4点完成的作业,按folder分组

```python
prefixs_2 = ('SYS_FW','SYS_FS','FW_LD','LD_TDW','STG_V','RT_RST','FW_TS','STG_C','PDM_P','EX_P','EXM_P',\
           'PDM_C','PDM_F','EXZ_P','NDS_P','NDS_C','EX_T','EXM_T','EXZ_T','EX_A','EXM_A','BRTL_C',\
           'RF_RST','FW_TB','BRTL_P','TS_HEES','EXZ_A','TS_HOSM','HEES_ETL1','EXQ_P','TRM_HCXI','TRZ_HCXI','TR_HCXI','BCOR_P')
def repalce_jobname_head(job_head):
    if job_head in prefixs_2:
        return job_head
    else:
        return 'head_2_other'
    
allday_list =[]
allday_index = []


    
for day in d_list:
    print(day,end =' ')
    everyday_list=[]
    
    #day = '181028'
    endt_th = int('20'+ day +'040000')
    
    df = pd.read_csv(r'E:\jupyter\CTM\data_src\ajob_0\ajob_%s_0.csv'%day,low_memory=False)
    
    df = df[df['end_time'].notnull()]
    df = df[df['end_time']<endt_th]
    
    df['job_prefix'] = df['job_name'].apply(lambda x:'_'.join(x.split("_")[0:2]))
    df['job_prefix'] = df['job_prefix'].apply(repalce_jobname_head)
    
    
    
    for hd_2 in prefixs_2:
        everyday_list.append(df[df.job_prefix==hd_2].order_id.count())
    
    
    everyday_list.append(df['order_id'].count())

    allday_list.append(everyday_list)
    allday_index.append(day)
    
colms = [cl+'_finsih0_4_count' for cl in prefixs_2]+['all_finsat_0_4_count'] 
fnish_at_0_4_countdf = pd.DataFrame(allday_list,columns=colms)
fnish_at_0_4_countdf.index = allday_index
print(fnish_at_0_4_countdf.shape)
```
