### ajob表进一步处理，加入分组统计信息(end_time,start_time)

这里截取的4点之前的数据进行统计的。
如果按天进行多个模型训练，更改截取时间.

```python
start = time.time() 

for day in d_list:
    
    #day = '180602'
#     dat1 = datetime.datetime.strptime('20'+day+'000000','%Y%m%d%H%M%S')    
    day_i  = day_all.index(day)
#     day_int = int(day)
#     day_l = '20'+day

#     if day == '180709':
#         thred_hour = '100000'
#     elif day == '180710':
#         thred_hour = '090000'
#     elif day == '180711':
#         thred_hour = '080000'
#     elif day == '180712':
#         thred_hour = '070000'
#     elif day == '180713':
#         thred_hour = '060000'
#     elif day == '180714':
#         thred_hour = '050000'
#     elif day == '180715':
#         thred_hour = '040000'

#     day_begain = int(day_l+'000000')
#     day_4_hour = int(day_l+'040000')  #4点开始

    train_src = pd.read_csv(r'E:\jupyter\CTM\data_processed\ajob\ajob_%s_1.csv'%day,low_memory=False)


    #据预处理

    train_src = train_src[['order_id','job_name','server','priority','nodegroup','job_name_head','order_table']]


    #--------------------------------------------------------------------------------


    #获取作业前n天进度统计情况，按分组统计  4个分组
    def get_n_dago_group_st(group_cls,n,pri_list):
        day_s = int(day_all[day_i - n]+'00')
        day_e = int(day_all[day_i - n]+'03')
        group_dict = group_cls.ix[day_s:day_e,:].apply(lambda x: x.sum()).to_dict()

        return pri_list.apply(lambda x:group_dict[x])

    def get_td_group_st(group_cls,hour,pri_list):
        day_ind = int(day + hour)
        return pri_list.apply(lambda x:group_cls[x][day_ind])

    #priority
    for k in range(0,15):
        train_src['day_%d_bef_endrt_by_pri'%k] = get_n_dago_group_st(end_by_priority,k,train_src['priority'])
    day1_7ago_endrt_by_pri = ['day_%d_bef_endrt_by_pri'%k for k in range(1,8)]
    day1_14ago_0endrt_by_pri=['day_%d_bef_endrt_by_pri'%k for k in range(1,15)]
    for f in f_sum_std:
        train_src['day1_7ago_endrt_by_pri_%s'%f] = train_src[day1_7ago_endrt_by_pri].apply(f, axis=1)
        train_src['day1_14ago_endrt_by_pri_%s'%f] = train_src[day1_14ago_0endrt_by_pri].apply(f, axis=1)
        if f != 'std':
            train_src['day1_7ago_endrt_bypri_sub_%s'%f] = train_src[['day_0_bef_endrt_by_pri','day1_7ago_endrt_by_pri_%s'%f]].apply(substrc, axis=1)
            train_src['day1_14ago_endrt_bypri_sub_%s'%f] = train_src[['day_0_bef_endrt_by_pri','day1_14ago_endrt_by_pri_%s'%f]].apply(substrc, axis=1)

            del train_src['day1_7ago_endrt_by_pri_%s'%f]
            del train_src['day1_14ago_endrt_by_pri_%s'%f]

    for j in range(1,15):
        clm = 'day_%d_bef_endrt_by_pri'%j
        del train_src[clm]    

    #nodegroup
    for k in range(0,15):
        train_src['day_%d_bef_endrt_by_node'%k] = get_n_dago_group_st(end_by_nodegroup,k,train_src['nodegroup'])
    day1_7ago_endrt_by_node = ['day_%d_bef_endrt_by_node'%k for k in range(1,8)]
    day1_14ago_0endrt_by_node=['day_%d_bef_endrt_by_node'%k for k in range(1,15)]
    for f in f_sum_std:
        train_src['day1_7ago_endrt_by_node_%s'%f] = train_src[day1_7ago_endrt_by_node].apply(f, axis=1)
        train_src['day1_14ago_endrt_by_node_%s'%f] = train_src[day1_14ago_0endrt_by_node].apply(f, axis=1)

        if f != 'std':
            train_src['day1_7ago_endrt_bynode_sub_%s'%f] = train_src[['day_0_bef_endrt_by_node','day1_7ago_endrt_by_node_%s'%f]].apply(substrc, axis=1)
            train_src['day1_14ago_endrt_bynode_sub_%s'%f] = train_src[['day_0_bef_endrt_by_node','day1_14ago_endrt_by_node_%s'%f]].apply(substrc, axis=1)

            del train_src['day1_7ago_endrt_by_node_%s'%f]
            del train_src['day1_14ago_endrt_by_node_%s'%f]
    for j in range(1,15):
        clm = 'day_%d_bef_endrt_by_node'%j
        del train_src[clm]

    #folder  order_table
    for k in range(0,15):
        train_src['day_%d_bef_endrt_by_folder'%k] = get_n_dago_group_st(end_by_folder,k,train_src['order_table'])
    day1_7ago_endrt_by_folder = ['day_%d_bef_endrt_by_folder'%k for k in range(1,8)]
    day1_14ago_0endrt_by_folder=['day_%d_bef_endrt_by_folder'%k for k in range(1,15)]
    for f in f_sum_std:
        train_src['day1_7ago_endrt_by_folder_%s'%f] = train_src[day1_7ago_endrt_by_folder].apply(f, axis=1)
        train_src['day1_14ago_endrt_by_folder_%s'%f] = train_src[day1_14ago_0endrt_by_folder].apply(f, axis=1)

        if f != 'std':
            train_src['day1_7ago_endrt_byfolder_sub_%s'%f] = train_src[['day_0_bef_endrt_by_folder','day1_7ago_endrt_by_folder_%s'%f]].apply(substrc, axis=1)
            train_src['day1_14ago_endrt_byfolder_sub_%s'%f] = train_src[['day_0_bef_endrt_by_folder','day1_14ago_endrt_by_folder_%s'%f]].apply(substrc, axis=1)

            del train_src['day1_7ago_endrt_by_folder_%s'%f]
            del train_src['day1_14ago_endrt_by_folder_%s'%f]
    for j in range(1,15):
        clm = 'day_%d_bef_endrt_by_folder'%j
        del train_src[clm] 

    #prefix    
    for k in range(0,15):
        train_src['day_%d_bef_endrt_by_prefix'%k] = get_n_dago_group_st(end_by_prefix,k,train_src['job_name_head'])
    day1_7ago_endrt_by_folder = ['day_%d_bef_endrt_by_prefix'%k for k in range(1,8)]
    day1_14ago_0endrt_by_folder=['day_%d_bef_endrt_by_prefix'%k for k in range(1,15)]
    for f in f_sum_std:
        train_src['day1_7ago_endrt_by_prefix_%s'%f] = train_src[day1_7ago_endrt_by_folder].apply(f, axis=1)
        train_src['day1_14ago_endrt_by_prefix_%s'%f] = train_src[day1_14ago_0endrt_by_folder].apply(f, axis=1)

        if f != 'std':
            train_src['day1_7ago_endrt_byprefix_sub_%s'%f] = train_src[['day_0_bef_endrt_by_prefix','day1_7ago_endrt_by_prefix_%s'%f]].apply(substrc, axis=1)
            train_src['day1_14ago_endrt_byprefix_sub_%s'%f] = train_src[['day_0_bef_endrt_by_prefix','day1_14ago_endrt_by_prefix_%s'%f]].apply(substrc, axis=1)

            del train_src['day1_7ago_endrt_by_prefix_%s'%f]
            del train_src['day1_14ago_endrt_by_prefix_%s'%f]
    for j in range(1,15):
        clm = 'day_%d_bef_endrt_by_prefix'%j
        del train_src[clm]

    #------------------------------------------------------------------------------------------------



    del_clm = ('priority','nodegroup','job_name_head','order_table')

    for cl in del_clm:
        del train_src[cl]


    #train_src.drop_duplicates('job_name', keep='last', inplace=True)


    train_src = train_src.round(3)
    train_src.to_csv(r'E:\jupyter\CTM\data_processed\ajob2\ajob_%s_2.csv'%(day),index=False)


    print(day,end=' ')
    end = time.time()
    print('    %0.2f minute'%((end-start)/60)) 
```
